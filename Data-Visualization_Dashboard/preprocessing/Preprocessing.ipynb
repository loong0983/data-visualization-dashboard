```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
```

```python
remain_columns = ['Country Name', 'Country Code'] + [str(i) for i in range(1991, 2019)]
```

```python
not_country = ['ARB', 'CEB', 'CSS', 'EAP', 'EAR', 'EAS', 'ECA', 'ECS', 
               'EMU', 'EUU', 'FCS', 'HIC', 'HPC', 'IBD', 'IBT', 'IDA', 
               'IDB', 'IDX', 'INX', 'LAC', 'LCN', 'LDC', 'LIC', 'LMC', 
               'LMY', 'LTE', 'MEA', 'MIC', 'MNA', 'NAC', 'OED', 'OSS',
               'PRE', 'PSS', 'PST', 'SAS', 'SSA', 'SSF', 'SST', 'TEA',
               'TEC', 'TLA', 'TMN', 'TSA', 'TSS', 'UMC', 'WLD', '']
```

```python
def preprocess_df(df):
    dropped_country = []
    
    df_new = df[remain_columns]
    df_copy = df_new.copy()
    for idx, row in df_new.iterrows():
        data = row[[str(i) for i in range(1991, 2019)]]
        if data.isna().any() == True:
            dropped_country.append(row['Country Name'])
            df_copy.drop(idx, inplace=True)
    return df_copy, dropped_country
```

### Labour Force by Continent

```python
os.listdir('./dataset/Labour Force Male')
```

```python
labour_female = pd.read_csv('./dataset/Labour Force Female/API_SL.TLF.CACT.FE.ZS_DS2_en_csv_v2_821018.csv', skiprows=4)
labour_male = pd.read_csv('./dataset/Labour Force Male/API_SL.TLF.CACT.MA.ZS_DS2_en_csv_v2_821025.csv', skiprows=4)
```

```python
labour_female, d1 = preprocess_df(labour_female)
labour_male, d2 = preprocess_df(labour_male)
```

```python
d1 == d2, len(d1), len(d2)
```

```python
print(labour_female[labour_female.isna().any(axis=1)])
print(labour_male[labour_male.isna().any(axis=1)])
```

```python
labour_female_copy = labour_female.copy()
for idx, row in labour_female.iterrows():
    if row['Country Code'] in not_country:
        labour_female_copy.drop(idx, inplace=True)
labour_female = labour_female_copy

labour_male_copy = labour_male.copy()
for idx, row in labour_male.iterrows():
    if row['Country Code'] in not_country:
        labour_male_copy.drop(idx, inplace=True)
labour_male = labour_male_copy
```

```python
labour_male_mean = labour_male.mean(axis=1)
labour_male_mean.name = 'Labour Force Male'
labour_female_mean = labour_female.mean(axis=1)
labour_female_mean.name = 'Labour Force Female'
```

```python
labour_male['Country Name']
```

```python
labour_by_gender = pd.DataFrame([labour_male['Country Name'], labour_male['Country Code'], labour_male_mean, labour_female_mean]).transpose()
labour_by_gender
```

```python
continent = pd.read_csv('./dataset/Labour Force Male/Metadata_Country_API_SL.TLF.CACT.MA.ZS_DS2_en_csv_v2_821025.csv')
```

```python
labour_female['Region'] = labour_female['Country Code'].apply(lambda x: continent[continent['Country Code'] == x].Region.item())
labour_male['Region'] = labour_male['Country Code'].apply(lambda x: continent[continent['Country Code'] == x].Region.item())
```

```python
labour_male.columns
```

```python
labour_male.Region.value_counts()
```

```python
labour_male_sub = labour_male[['Region'] + [str(i) for i in range(1991, 2019)]]
labour_female_sub = labour_female[['Region'] + [str(i) for i in range(1991, 2019)]]
```

```python
labour_male_sub = labour_male_sub.groupby('Region').mean()
labour_female_sub = labour_female_sub.groupby('Region').mean()
```

```python
labour_male_sub.to_csv('./Preprocessed data/labour_force_male_region_year.csv')
labour_female_sub.to_csv('./Preprocessed data/labour_force_female_region_year.csv')
```

```python
labour_male_mean = labour_male_sub.mean(axis=1)
labour_male_mean.name = 'Male'
labour_male_mean
```

```python
labour_female_mean = labour_female_sub.mean(axis=1)
labour_female_mean.name = 'Female'
labour_female_mean
```

```python
labour = pd.concat([labour_male_mean, labour_female_mean], axis=1)
labour
```

```python
male_norm = []
female_norm = []
for idx, row in labour.iterrows():
    denominator = row.Male + row.Female
    male_norm.append(row.Male / denominator * 100)
    female_norm.append(row.Female / denominator * 100)
```

```python
labour['Male Normalized'] = male_norm
labour['Female Normalized'] = female_norm
labour
```

```python
labour.to_csv('Preprocessed data/labour_force.csv')
```
